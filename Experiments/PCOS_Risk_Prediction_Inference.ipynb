```python
from tensorflow.keras.models import load_model
import joblib
import pandas as pd

basic_model = load_model("saved_models/basic_ann_model.h5")
basic_scaler = joblib.load("saved_models/basic_ann_scaler.pkl")
basic_features = joblib.load("saved_models/basic_ann_features.pkl")

new_data = pd.DataFrame([{
    "Age (yrs)": 25,
    "BMI": 28.5,
    "Cycle(R/I)": 1,
    "Pimples(Y/N)": 1,
    "hair growth(Y/N)": 1,
    "Hair loss(Y/N)": 0,
    "Weight gain(Y/N)": 1,
    "Skin darkening (Y/N)": 1,
    "Fast food (Y/N)": 1,
    "Reg.Exercise(Y/N)": 0
}])

new_data = new_data[basic_features]

new_scaled = basic_scaler.transform(new_data)

prob = basic_model.predict(new_scaled)[0][0]

print("PCOS Probability:", prob)
```
